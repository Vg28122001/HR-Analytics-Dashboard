```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file = pd.read_csv('HR_Analytics.csv')
```

```python
file.shape
```

```python
file.head()
```

```python
file.drop_duplicates(inplace = True)
file
```

```python
file['Attrition_Count'] = np.where(file['Attrition'] == 'Yes', 1, 0)
file
```

```python
Total_Attrition = file['Attrition_Count'].value_counts()[1]
Total_Attrition
```

#### Attrition By Education

```python
AbE = file.groupby('EducationField')['Attrition_Count'].sum()
AbE
```

```python
plt.pie(AbE, labels=AbE.index, autopct='%1.1f%%')
plt.title('Attrition by Education')
plt.show()
```

#### Attrition by Age Group

```python
AbAG = file.groupby('AgeGroup')['Attrition_Count'].sum()
AbAG = AbAG.sort_values(ascending = False)
AbAG
```

```python
AbAG.plot( kind='bar', stacked=True)
plt.title('Attrition by Age Group')
plt.show()
```

#### Attrition by Salary

```python
AbS = file.groupby('SalarySlab')['Attrition_Count'].sum()
AbS = AbS.sort_values(ascending = False)
AbS
```

```python
AbS.plot( kind='barh', stacked=True)
plt.title('Attrition by Salary')
plt.show()
```

#### Attrition by Job Role

```python
AbJR = file.groupby('JobRole')['Attrition_Count'].sum()
AbJR = AbJR.sort_values(ascending = False)
AbJR
```

```python
AbJR.plot( kind='barh', stacked=True)
plt.title('Attrition by Job Role')
plt.show()
```

#### Attrition by Year at Company

```python
AbYC = file.groupby('YearsAtCompany')['Attrition_Count'].sum()
AbYC = AbYC.sort_index()
AbYC
```

```python
AbYC.plot( kind='area', stacked=True)
plt.title('Attrition by Year at Company')
plt.show()
```
